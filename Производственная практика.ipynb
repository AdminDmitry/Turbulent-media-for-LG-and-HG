<h2>Институт систем обработки изображений (ИСОИ) РАН</h2>
<p>Моделирование прохождения мод Лагерра-Гаусса и Эрмита-Гаусса в турбулентных средах с различными показателями турбулентности</p>

```python
from LightPipes import *
import matplotlib.pyplot as plt
from math import pi
from mpl_toolkits.axes_grid1 import make_axes_locatable

def Fres(F, mp):
    z = 50*mm
    Fn = F
    for i in range(60):
        Fn = RandomPhase(Fn, maxPhase=mp) # добавление случайного фазового сдвига для моделирования турбулентности
        Fn = Fresnel(Fn, z/60) # распространение на глубину z/60
    return Intensity(Fn), Phase(Fn)

wavelength = 500*nm
size = 15*mm
N = 501
w0 = 3*mm

F = Begin(size, wavelength, N)
F = GaussBeam(F, w0, LG=True, n=2, m=2)

P = Phase(F) # Фаза изначального пучка
I = Intensity(F) #Интенсивность изначального пучка

#Интенсивности и фазы пучка при прохождении турбулентных сред различной силы
I1, P1 = Fres(F, 0.1)
I2, P2 = Fres(F, 0.3)
I3, P3 = Fres(F, 0.5)

fig, axs = plt.subplots(nrows=2, ncols=4,figsize=(10.0,5.0))

fig.suptitle('Прохождение $ЛГ_{22}$ через турбулентную среду')
fig.subplots_adjust(hspace=0.12)


img1 = axs[0][0].imshow(P, cmap='jet'); axs[0][0].axis('off')
axs[1][0].imshow(I, cmap='jet'); axs[1][0].axis('off')

axs[0][1].imshow(P1, cmap='jet'); axs[0][1].axis('off')
axs[1][1].imshow(I1, cmap='jet'); axs[1][1].axis('off')

axs[0][2].imshow(P2, cmap='jet'); axs[0][2].axis('off')
axs[1][2].imshow(I2, cmap='jet'); axs[1][2].axis('off')

axs[0][3].imshow(P3, cmap='jet'); axs[0][3].axis('off')
axs[1][3].imshow(I3, cmap='jet'); axs[1][3].axis('off')


divider = make_axes_locatable(axs[0][3])
cax = divider.append_axes("right", size="5%", pad=0.05)
cbar=plt.colorbar(img1, cax=cax, ticks=[0, PI, 2*PI], orientation='vertical')
cbar.ax.set_yticklabels(['0', '$\pi$', '$2\pi$'])
```

```python
from LightPipes import *
import matplotlib.pyplot as plt
from math import pi
from mpl_toolkits.axes_grid1 import make_axes_locatable

def Fres(F, z):
    mp = 0.1
    Fn = F
    for i in range(60):
        Fn = RandomPhase(Fn, maxPhase=mp) # добавление случайного фазового сдвига для моделирования турбулентности
        Fn = Fresnel(Fn, z/60) # распространение на глубину delta_z
    return Intensity(Fn), Phase(Fn)

wavelength = 500*nm
size = 15*mm
N = 501
w0 = 3*mm

F = Begin(size, wavelength, N)
F = GaussBeam(F, w0, LG=False, n=2, m=2)

P = Phase(F) # Получение интенсивности пучка
I = Intensity(F)

I1, P1 = Fres(F, 50*mm)
I2, P2 = Fres(F, 100*mm)
I3, P3 = Fres(F, 150*mm)

fig, axs = plt.subplots(nrows=2, ncols=4,figsize=(10.0,5.0))

fig.suptitle('Прохождение $ЭГ_{22}$ через турбулентную среду')
fig.subplots_adjust(hspace=0.12)


# Расчет поля после прохождения через турбулентную среду

img1 = axs[0][0].imshow(P, cmap='jet'); axs[0][0].axis('off')
axs[1][0].imshow(I, cmap='jet'); axs[1][0].axis('off')

axs[0][1].imshow(P1, cmap='jet'); axs[0][1].axis('off')
axs[1][1].imshow(I1, cmap='jet'); axs[1][1].axis('off')

axs[0][2].imshow(P2, cmap='jet'); axs[0][2].axis('off')
axs[1][2].imshow(I2, cmap='jet'); axs[1][2].axis('off')

axs[0][3].imshow(P3, cmap='jet'); axs[0][3].axis('off')
axs[1][3].imshow(I3, cmap='jet'); axs[1][3].axis('off')


divider = make_axes_locatable(axs[0][3])
cax = divider.append_axes("right", size="5%", pad=0.05)
cbar=plt.colorbar(img1, cax=cax, ticks=[0, PI, 2*PI], orientation='vertical')
cbar.ax.set_yticklabels(['0', '$\pi$', '$2\pi$'])
```
